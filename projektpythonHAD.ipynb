```python
import pandas as pd
import math
```

```python
df_collection_nekrizove = pd.DataFrame(columns=["Week", "Pricing List", "Reputation", "Business Conditions", "Pillar3 related", "Pillar3 disclosure requirements"])

# Naplnenie stĺpca 'Week' v dátovom rámci
df_collection_nekrizove['Week'] = range(53)

# Naplnenie zvyšných stĺpcov hodnotami 0
df_collection_nekrizove[["Pricing List", "Reputation", "Business Conditions", "Pillar3 related", "Pillar3 disclosure requirements"]] = 0

# Výpočet a naplnenie hodnôt pre zvyšné stĺpce v cykloch
for col in df_collection_nekrizove.columns[1:]:
    for j in range(53):
        col_index = df_collection_nekrizove.columns.get_loc(col)
        intercept = float(data.loc[data['Effect'] == col, 'Intercept'])
        week = float(data.loc[data['Effect'] == col, 'Week'])
        week_sq = float(data.loc[data['Effect'] == col, 'Week Square'])
        week_cub = float(data.loc[data['Effect'] == col, 'weekCubic'])
        week_cub = float(data.loc[data['Effect'] == col, 'weekCubic'])
        #crisis pridane neskor
        # crisis = float(data.loc(data['crisis'] == col, 'crisis')
        df_collection_nekrizove.iloc[j, col_index] = intercept + (j * week) + (j * j * week_sq) + (j * j * j * week_cub)
```

```python
df_collection_nekrizove
```

```python
df_collection_nekrizove['ref'] = 0

for i in range(len(df_collection_nekrizove)):
    df_collection_nekrizove['ref'][i] = 1 / (1 + math.exp(df_collection_nekrizove['Pricing List'][i]) + math.exp(df_collection_nekrizove['Reputation'][i]) + 
                                             math.exp(df_collection_nekrizove['Business Conditions'][i]) + math.exp(df_collection_nekrizove['Pillar3 related'][i]) + 
                                             math.exp(df_collection_nekrizove['Pillar3 disclosure requirements'][i]))
df_collection_nekrizove
```

```python
df_collection_nekrizove['pravd_Pricing List'] = 0
df_collection_nekrizove['pravd_Reputation'] = 0
df_collection_nekrizove['pravd_Business Conditions'] = 0
df_collection_nekrizove['pravd_Pillar3 related'] = 0
df_collection_nekrizove['pravd_Pillar3 disclosure requirements'] = 0
```

```python
for i in range(len(df_collection_nekrizove)):
    df_collection_nekrizove['pravd_Pricing List'][i] = math.exp(df_collection_nekrizove['Pricing List'][i]) * df_collection_nekrizove['ref'][i]
    df_collection_nekrizove['pravd_Reputation'][i] = math.exp(df_collection_nekrizove['Reputation'][i]) * df_collection_nekrizove['ref'][i]
    df_collection_nekrizove['pravd_Business Conditions'][i] = math.exp(df_collection_nekrizove['Business Conditions'][i]) * df_collection_nekrizove['ref'][i]
    df_collection_nekrizove['pravd_Pillar3 related'][i] = math.exp(df_collection_nekrizove['Pillar3 related'][i]) * df_collection_nekrizove['ref'][i]
    df_collection_nekrizove['pravd_Pillar3 disclosure requirements'][i] = math.exp(df_collection_nekrizove['Pillar3 disclosure requirements'][i]) * df_collection_nekrizove['ref'][i]

df_collection_nekrizove
```

```python
import statistics
import plotly.graph_objects as px
import plotly.graph_objects as go
```

```python
fig = go.Figure()
fig.update_layout(showlegend=False)


line_styles = [
    dict(dashdot="dot", line_color="blue"),
    dict(dashdot="dashdot", line_color="green"),
    dict(dashdot="dash", line_color="red"),
    dict(dashdot="solid", line_color="orange"),
    dict(dashdot="solid", line_color="yellow")
]
```

```python
columns=["Pricing List", "Reputation", "Business Conditions", "Pillar3 related", "Pillar3 disclosure requirements"]

rozdiely = pd.DataFrame(columns=columns)
for i in range (53):
    rozdiely.loc[i] = [0,0,0,0,0]

for i in range (53):
    rozdiely['Pricing List'][i] = crosstabs["Pricing List"] - teo_pocetnosti["Pricing List"]
    rozdiely['Reputation'][i] = crosstabs["Reputation"] - teo_pocetnosti["Reputation"]
    rozdiely['Business Conditions'][i] = crosstabs["Business Conditions"] - teo_pocetnosti["Business Conditions"]
    rozdiely['Pillar3 related'][i] = crosstabs["Pillar3 related"] - teo_pocetnosti["Pillar3 related"]
    rozdiely['Pillar3 disclosure requirements'][i] = crosstabs["Pillar3 disclosure requirements"] - teo_pocetnosti["Pillar3 disclosure requirements"]

rozdiely
```

```python
names = ["Pricing List", "Reputation", "Business Conditions", "Pillar3 related", "Pillar3 disclosure requirements"]
x_sur = list(range(53))


fig.add_trace(go.Scatter(x=x_sur, y=rozdiely['Pricing List'], mode='lines', marker_color="red", name=names[0]))
mean = statistics.mean(rozdiely['Pricing List'])
stdev = statistics.stdev(rozdiely['Pricing List'])
fig.add_shape(type="line", x0=0, x1=52, y0=mean + 2 * stdev, y1=mean + 2 * stdev, line=dict(dash="dashdot", color="red"))
fig.add_shape(type="line", x0=0, x1=52, y0=mean - 2 * stdev, y1=mean - 2 * stdev, line=dict(dash="dashdot", color="red"))

fig.add_trace(go.Scatter(x=x_sur, y=rozdiely['Reputation'], mode='lines', marker_color="blue", name=names[1]))
mean = statistics.mean(rozdiely['Reputation'])
stdev = statistics.stdev(rozdiely['Reputation'])
fig.add_hline(y=mean + 2 * stdev, line=dict(dash="dashdot", color="blue"))
fig.add_hline(y=mean - 2 * stdev, line=dict(dash="dashdot", color="blue"))

fig.add_trace(go.Scatter(x=x_sur, y=rozdiely['Business Conditions'], mode='lines', marker_color="green", name=names[2]))
mean = statistics.mean(rozdiely['Business Conditions'])
stdev = statistics.stdev(rozdiely['Business Conditions'])
fig.add_hline(y=mean + 2 * stdev, line=dict(dash="dashdot", color="green"))
fig.add_hline(y=mean - 2 * stdev, line=dict(dash="dashdot", color="green"))

fig.add_trace(go.Scatter(x=x_sur, y=rozdiely['Pillar3 related'], mode='lines', marker_color="orange", name=names[3]))
mean = statistics.mean(rozdiely['Pillar3 related'])
stdev = statistics.stdev(rozdiely['Pillar3 related'])
fig.add_hline(y=mean + 2 * stdev, line=dict(dash="dashdot", color="orange"))
fig.add_hline(y=mean - 2 * stdev, line=dict(dash="dashdot", color="orange"))

fig.add_trace(go.Scatter(x=x_sur, y=rozdiely['Pillar3 disclosure requirements'], mode='lines', marker_color="yellow", name=names[4]))
mean = statistics.mean(rozdiely['Pillar3 disclosure requirements'])
stdev = statistics.stdev(rozdiely['Pillar3 disclosure requirements'])
fig.add_hline(y=mean + 2 * stdev, line=dict(dash="dashdot", color="yellow"))
fig.add_hline(y=mean - 2 * stdev, line=dict(dash="dashdot", color="yellow"))

fig.show()
```
